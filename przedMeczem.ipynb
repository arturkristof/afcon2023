```python
def print_next_game(df, next_game_num):
    game = df.loc[next_game_num]

    country1 = game["country1"]
    country2 = game["country2"]


    print(f"Next game (#{next_game_num}): {country1} - {country2}")
    for name in names:
        res1 = game[f"{name}1"]
        res2 = game[f"{name}2"]
        
        left_win = False
        right_win = False
        
        is_asterisk = False
        # read res1
        try:
            res1 = int(res1)
        except:
            if("*" in res1):
                left_win = True
                is_asterisk = True
            res1 = int("".join([c for c in res1 if c != "*"]))
        # read res2
        try:
            res2 = int(res2)
        except:
            if("*" in res2):
                right_win = True
                is_asterisk = True
            res2 = int("".join([c for c in res2 if c != "*"]))
            
        if((not left_win) and (not right_win)):
            left_win = res1 > res2
            right_win = res2 > res1
        
        line = f"{name} - {res1}:{res2}"
        if(is_asterisk == True):
            country = country1 if left_win else country2
            line += f" ({country})"
            
        print(line)
    print("")

    
res_df = df.iloc[1:49, :].append(df.iloc[51:59, :]).append(df.iloc[61:65, :]).append(df.iloc[67:69, :]).append(df.iloc[71, :]).append(df.iloc[74, :])
res_df = res_df.reset_index(drop=True)
#res_df.tail(10)
next_game_num = res_df['score1'].last_valid_index()

print_next_game(res_df, next_game_num)
```

```python
def print_ordered_dict(results_dict, title, reverse=False):
    res = pd.Series(results_dict)
    res = res.sort_values(ascending=reverse)
    
    print(f"{title}:")
    
    for name in res.index:
        print(f"{name} - {res[name]}")
```

## Goals off per team per match

```python
def res_mapping(x):
    if(isinstance(x, float)):
        return int(x)
    return int("".join([c for c in x if c != "*"]))

score1 = res_df["score1"]
score2 = res_df["score2"]

last_game_idx = score1.last_valid_index() - 2

score1 = score1[:last_game_idx].astype(int)
score2 = score2[:last_game_idx].astype(int)

goal_diff_dict = {}

for name in names:
    guess1 = res_df[f"{name}1"]
    guess2 = res_df[f"{name}2"]


    guess1 = guess1[:last_game_idx].map(res_mapping)
    guess2 = guess2[:last_game_idx].map(res_mapping)
    
    diff1 = (guess1 - score1).abs()
    diff2 = (guess2 - score2).abs()
    
    goal_diff_total = (diff1 + diff2).sum()
    
    goal_diff_dict[name] = round(goal_diff_total/score1.size,2)

# next_game_num = res_df['score1'].last_valid_index() + 1
# test1.astype(int) - test2.astype(int)

print_ordered_dict(goal_diff_dict, "Roznica goli na mecz", reverse=True)
```

## Losing streak

```python
from itertools import groupby

def get_longest_pizda_streak(points):
    lst = []
    for n,c in groupby(points):
       num,count = n,sum(1 for i in c)
       lst.append((num,count))

    streak = max([y for x,y in lst if x == '0'])
    return streak
```

```python
next_game_num = df['score1'].last_valid_index()
played_games = df[:next_game_num]

pizda_streak_dict = {f"{name}" : get_longest_pizda_streak(played_games[f"{name}Pkt"]) for name in names}

pizda_streak = pd.Series(pizda_streak_dict)
pizda_streak = pizda_streak.sort_values(ascending=False)

print("Pizda streak:")
for name in pizda_streak.index:
    name_points = played_games[f"{name}Pkt"][-pizda_streak[name]:]
    is_ongoing = True
    for x in name_points:
        if(x != '0'):
            is_ongoing = False
    line = f"{name} - {pizda_streak[name]}"
    if(is_ongoing):
        line += " (ongoing)"
    print(line)
```
